```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Year": [2000, 2005, 2010, 2015, 2020],
    "Avg_Temperature_C": [14.3, 14.6, 14.8, 15.1, 15.5],
    "CO2_ppm": [369, 379, 389, 400, 414],
    "Rainfall_mm": [1100, 1050, 1150, 1200, 1180]
}
    
df = pd.DataFrame(data)
df

print("Dataset:")
print(df)

print("\nSummary Statistics:")
print(df.describe())


plt.plot(df["Year"], df["Avg_Temperature_C"], marker='o', label="Avg Temp (Â°C)")
plt.plot(df["Year"], df["CO2_ppm"]/30, marker='s', label="CO2 (scaled)")
plt.title("Climate Trends Over Time")
plt.xlabel("Year")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()

print("\nCorrelation between temperature and CO2:")
print(df[["Avg_Temperature_C", "CO2_ppm"]].corr())
```
