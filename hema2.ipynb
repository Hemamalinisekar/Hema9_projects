```python
import pandas as pd
df = pd.read_csv("/home/umashankar/Downloads/sales_data.csv")
print(df.head())
df.info()       
df.describe()    
df.isnull().sum() 
df = df.drop_duplicates()
df['Price'] = df['Price'].astype(float)
df['Total'] = df['Price'] * df['Quantity'] 
category_sales = df.groupby('Category')['Total'].sum()
print(category_sales)
region_sales = df.groupby('Region')['Total'].sum()
print(region_sales)
top_product = df.groupby('Product')['Total'].sum().sort_values(ascending=False).head(1)
print(top_product)
avg_quantity = df.groupby('Category')['Quantity'].mean()
print(avg_quantity)
import matplotlib.pyplot as plt
category_sales.plot(kind='bar', title='Total Sales by Category')
plt.show()
region_sales.plot(kind='pie', title='Sales by Region', autopct='%1.1f%%')
plt.show()
df.to_csv("cleaned_sales_data.csv", index=False)
```
